```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/Dataset .csv")

# Display first 5 rows
df.head()
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Fill missing values in the 'Cuisines' column with "Unknown"
df["Cuisines"].fillna("Unknown", inplace=True)

# Verify if missing values are handled
df.isnull().sum()
```

```python
# Drop unnecessary columns
columns_to_drop = ["Country Code", "Address", "Locality", "Locality Verbose", "Currency", "Switch to order menu"]
df.drop(columns=columns_to_drop, inplace=True)

# Display the updated dataset columns
df.columns
```

```python
# Convert 'Yes' to 1 and 'No' to 0 in categorical columns
df["Has Table booking"] = df["Has Table booking"].map({"Yes": 1, "No": 0})
df["Has Online delivery"] = df["Has Online delivery"].map({"Yes": 1, "No": 0})
df["Is delivering now"] = df["Is delivering now"].map({"Yes": 1, "No": 0})

# Display the first few rows to verify changes
df.head()
```

```python
# Function to recommend restaurants based on user preferences
def recommend_restaurants(cuisine, max_price, min_rating):
    # Filter restaurants based on criteria
    filtered_df = df[
        (df["Cuisines"].str.contains(cuisine, case=False, na=False)) &
        (df["Average Cost for two"] <= max_price) &
        (df["Aggregate rating"] >= min_rating)
    ]

    # Sort by highest rating
    recommended = filtered_df.sort_values(by="Aggregate rating", ascending=False)

    return recommended[["Restaurant Name", "Cuisines", "Average Cost for two", "Aggregate rating"]].head(10)

# Example: Recommend restaurants with 'Italian' cuisine, max price 500, and rating 4.0+
recommend_restaurants("Italian", 500, 4.0)
```

```python
# Test 1: Recommend Chinese restaurants under â‚¹300 with rating 4.0+
recommend_restaurants("Chinese", 300, 4.0)
# Test 2: Recommend Indian restaurants under â‚¹600 with rating 3.5+
recommend_restaurants("Indian", 600, 3.5)
# Test 3: Recommend Fast Food restaurants under â‚¹250 with rating 4.2+
recommend_restaurants("Fast Food", 250, 4.2)
```

```python
# Get recommendations for a sample preference (Indian cuisine, â‚¹500 budget, 4.0+ rating)
recommendations = recommend_restaurants("Indian", 500, 4.0)

# Save the recommendations to a CSV file
recommendations.to_csv("/content/Restaurant_Recommendations.csv", index=False)

print("File saved successfully! âœ…")
```
